```python
import pandas as pd
import numpy as np

# ================================
# 1. Load Dataset
# ================================
df = pd.read_csv("dataset.csv")

# ================================
# 2. Drop rows with missing target
# ================================
df = df.dropna(subset=["price"])

# ================================
# 3. Fix data types
# ================================
numeric_cols = ["mileage", "cylinders", "doors"]
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# ================================
# 4. Handle missing values
# ================================

# --- Numeric ---
df["mileage"].fillna(df["mileage"].median())
df["cylinders"].fillna(df["cylinders"].mode()[0])
df["doors"].fillna(df["doors"].mode()[0])

# --- Categorical ---
df["engine"].fillna("Unknown")

categorical_cols = [
    "fuel",
    "transmission",
    "trim",
    "body",
    "exterior_color",
    "interior_color"
]

for col in categorical_cols:
    df[col].fillna(df[col].mode()[0])

# ================================
# 5. Drop unnecessary columns
# ================================
df.drop(columns=["description", "name"], inplace=True)

# ================================
# 6. Final validation
# ================================
print("Dataset Shape:", df.shape)
print("\nMissing Values:\n", df.isnull().sum())
print("\nData Types:\n", df.dtypes)
```

**VISUAL EDA**

```python
 # VISUAL EDA 

# 1. Price Distributi
plt.figure()
plt.hist(df["price"], bins=30)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()





```

```python
# 2. Log-Price Distribution (recommended)
plt.figure()
plt.hist(np.log1p(df["price"]), bins=30)
plt.title("Log Price Distribution")
plt.xlabel("Log(Price)")
plt.ylabel("Count")
plt.show()
```

```python
# 3. Price vs Mileage
plt.figure()
plt.scatter(df["mileage"], df["price"])
plt.title("Price vs Mileage")
plt.xlabel("Mileage")
plt.ylabel("Price")
plt.show()

```

```python
# 4. Price vs Year
plt.figure()
plt.scatter(df["year"], df["price"])
plt.title("Price vs Year")
plt.xlabel("Year")
plt.ylabel("Price")
plt.show()
```

```python
# 5. Boxplot: Price vs Body Type
plt.figure()
df.boxplot(column="price", by="body", rot=90)
plt.title("Price by Body Type")
plt.suptitle("")
plt.xlabel("Body Type")
plt.ylabel("Price")
plt.show()

```

```python
# 6. Correlation Heatmap (Numerical)
num_df = df[["price", "year", "mileage", "cylinders", "doors"]]
corr = num_df.corr()

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# FEATURE ENGINEERING

from datetime import datetime

# 1. Vehicle Age
current_year = datetime.now().year
df["vehicle_age"] = current_year - df["year"]

```

```python
# 2. Log transform target (for skewed price)
df["log_price"] = np.log1p(df["price"])
```

```python
# 3. Mileage bins
df["mileage_bin"] = pd.cut(
    df["mileage"],
    bins=[-1, 10000, 30000, 60000, 100000, np.inf],
    labels=["Very Low", "Low", "Medium", "High", "Very High"]
)
```

```python
# 4. Cylinders grouping
df["cylinder_group"] = df["cylinders"].apply(
    lambda x: "Low" if x <= 4 else ("Medium" if x <= 6 else "High")
)
```

```python
# 5. Simplify transmission
df["transmission_type"] = df["transmission"].fillna("Unknown").astype(str).apply(
    lambda x: "Automatic" if "Auto" in x else ("Manual" if "Manual" in x else "Other")
)
```

```python
# 6. Drop redundant columns
df.drop(columns=["year", "price"], inplace=True)

# ================================
# Final check
# ================================
df.head()
df.info()
```
